# Evaluation Metrics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
def get_cumulative_average(dataframe):
    '''Calculate the cumulative reward, average reward, and episodic reward after 70,000 timestep'''
    cum = pd.DataFrame()
    avg = pd.DataFrame()
    res = pd.DataFrame()
    for name in ['dqn', 'sarsa', 'ppo']:
        cols = [f'{name}_trad_reward', f'{name}_nontrad_reward', f'{name}_timestep']
        labels = ['Traditional Reward', 'Non-traditional Reward', 'Timestep']
        mod_df = dataframe[cols[:2]].copy()
        mod_df.columns = labels[:2]
        
        # cumulative reward and average reward
        cum[name.upper()] = mod_df.sum()
        avg[name.upper()] = mod_df.mean()
        
        # calculate episodic reward after 70,000 time steps
        mod_df2 = dataframe[cols].copy().cumsum()
        for index, row in mod_df2.iterrows():
            if row[f'{name}_timestep'] > 70000:
                row[f'{name}_timestep'] = 70000
                row.index = labels
                res[name.upper()] = row
                break

        # delete new lists created to free up memory
        del mod_df
        del mod_df2
  

    return cum, avg, res
```

```python
def plot_grid(df2, title = '', roundValues=False):
    '''Plot a graph of all the metrics'''

    df_melted = df2.reset_index().melt(id_vars='index', var_name='reward_type', value_name='reward')

    # Plot the graph using Seaborn
    sns.set(style="whitegrid")
    plt.figure(figsize=(7, 4))
    ax = sns.barplot(x='index', y='reward', hue='reward_type', data=df_melted)
    
    # Annotate bars with count values
    for p in ax.patches:
        height = f'{p.get_height():.2f}' if roundValues else  f'{int(p.get_height()):,}'
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')
    
    plt.xlabel('Reward type')
    plt.ylabel('Average Count' if "Average" in title else 'Count')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
def anova_test(dataframe):
    '''Test the statistical difference between performance of the algorithms'''
    data = pd.concat([dataframe['dqn_total_reward'], dataframe['sarsa_total_reward'], dataframe['ppo_total_reward']])
    groups = ['DQN Total Reward'] * 3000 + ['SARSA Total Reward'] * 3000 + ['PPO Total Reward'] * 3000

    # Create a DataFrame
    df4 = pd.DataFrame({'Data': data, 'Group': groups})

    model = ols('Data ~ Group', data=df4).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # # Perform Tukey's HSD post-hoc test
    posthoc = pairwise_tukeyhsd(df4['Data'], df4['Group'])

    print(anova_table)
    print(posthoc)
```
